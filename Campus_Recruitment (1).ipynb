##Splitting Data into Train and Test sets

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

##Label Encoder for Target
It is used to convert the labels of 'status' to numeric form.

```
from sklearn.preprocessing import LabelEncoder
```

```
le = LabelEncoder()
```

```
y_train = le.fit_transform(y_train)
```

```
y_train
```

```
y_test = le.fit_transform(y_test)
```

```
y_test
```

##Standardization
1. A constant range is created and values of all the columns are converted to be within this range .
2. It is used to get better efficiency.

```
from sklearn.preprocessing import StandardScaler
```

```
sc = StandardScaler()
```

```
X_train = sc.fit_transform(X_train)
X_test =  sc.transform(X_test)
```

##Train on different algorithms

```
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```
lr_cla = LogisticRegression()
knn_cla = KNeighborsClassifier()
dt_cla = DecisionTreeClassifier()
rf_cla = RandomForestClassifier()
lsvc_cla = SVC(kernel='linear')
ksvc_cla = SVC(kernel='rbf')
```

```
lr_cla.fit(X_train, y_train)
knn_cla.fit(X_train, y_train)
dt_cla.fit(X_train, y_train)
rf_cla.fit(X_train, y_train)
lsvc_cla.fit(X_train, y_train)
ksvc_cla.fit(X_train, y_train)
```

```
lr_pred = lr_cla.predict(X_test)
knn_pred = knn_cla.predict(X_test)
dt_pred = dt_cla.predict(X_test)
rf_pred = rf_cla.predict(X_test)
lsvc_pred = lsvc_cla.predict(X_test)
ksvc_pred = ksvc_cla.predict(X_test)
```

###Confusion Matirices

```
from sklearn.metrics import confusion_matrix
```

```
lr_c = confusion_matrix(y_test, lr_pred)
knn_c = confusion_matrix(y_test, knn_pred)
dt_c = confusion_matrix(y_test, dt_pred)
rf_c = confusion_matrix(y_test, rf_pred)
lsvc_c = confusion_matrix(y_test, lsvc_pred)
ksvc_c = confusion_matrix(y_test, ksvc_pred)
```
